My lane: Performance-based opportunity modeling (continuing the content-refresh thread from Notebooks 1 and 2).
I'm picking this because I already saw it work: a hand-written rule only got about 24% of the "fix this first" pages right, but a basic trained model got roughly 74% right on the same data. That gap is too big to ignore. I want to understand why the model picks different pages than the rule, using the numbers myself instead of trusting the notebook's summary.

Decision: which pages in the content backlog get reviewed and refreshed first.
Actor: a content/SEO person who can manually review maybe 50 pages a week, not all ~30,000.
Cost of a wrong call: if the ranking sends them to a low-value page, that's an hour not spent on a page actually losing traffic or rankings. Across a team, over months, that adds up to real wasted effort — and the pages that need attention keep sliding.

```python
reviewer_capacity_per_week = 50
print(f"Backlog size: {len(df)} pages")
print(f"Weeks to review everything at {reviewer_capacity_per_week}/week: {len(df) / reviewer_capacity_per_week:.0f}")
```

Looking at the ~30k pages in the starter dataset, three numbers stood out:

1. 37.5% of pages fall into the "low" impression tier — over a third of the backlog is barely getting seen in search at all.
2. The gap between the top and bottom quartile on average search position is 16.1 (Q1=6.2, Q3=22.3) — pages aren't clustered near one spot, they're spread across a wide performance range.
3. 25.7% of rows are missing word_count (and 71.5% are missing provider_used) — that's a data-quality issue my model will need to handle, likely by treating missingness itself as a signal rather than dropping those rows.

These numbers make the lane worth pursuing: the backlog is large enough, and spread out enough in performance, that even a modest improvement in ranking order saves real reviewer time.

```python
total_rows = len(df)

# X — % of pages in the lowest-performing bracket
print("Position tier breakdown (%):")
print((df['position_tier'].value_counts(normalize=True) * 100).round(1))

print("\nImpression tier breakdown (%):")
print((df['impression_tier'].value_counts(normalize=True) * 100).round(1))

# Y — quartile gap on a real performance metric
q1 = df['avg_position'].quantile(0.25)
q3 = df['avg_position'].quantile(0.75)
print(f"\navg_position: Q1={q1}, Q3={q3}, gap={q3 - q1:.1f}")

# Z — % missing on a column that matters for your lane
print(f"\n% rows missing word_count: {df['word_count'].isna().mean()*100:.1f}%")
print(f"% rows missing provider_used: {df['provider_used'].isna().mean()*100:.1f}%")
print(f"Total rows: {total_rows}")
```

What I can say: my findings are observed and measured directly from this anonymized sample — patterns in the data, and directional signals about which pages look higher-priority. Any output is decision-support: a ranked starting point for a human reviewer, not a final verdict.

What I can't say: I can't claim to predict Google's ranking algorithm, and I can't claim causation — that refreshing a page will increase traffic, only that pages with certain patterns have historically performed differently. I also can't generalize beyond this anonymized sample.
