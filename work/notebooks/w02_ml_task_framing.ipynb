## 1. My lane as an ML task (type)

This is a Classification task.
why? The goal is to categorize web pages into distinct buckets—specifically, identifying whether a page belongs to the "Needs Refresh" category or the "Leave Alone" category

```python
# Defining the distinct buckets for our Classification task
task_type = "Classification"
target_classes = ["Needs Refresh", "Leave Alone"]

print(f"ML Task Type: {task_type}")
print(f"The model will categorize pages into {len(target_classes)} discrete buckets: {target_classes}")
```

## 2. Target or proxy

The target is a binary proxy indicating a performance drop or an opportunity for improvement (e.g., a 1 for "Refresh" and 0 for "Leave alone"). Because we don't have a perfect "needs refresh" label from humans, we use historical traffic declines as our proxy for an update opportunity.

```python
# Defining the target proxy conceptually
target_column = "needs_refresh_proxy"
target_values = {1: "Refresh Opportunity", 0: "Leave Alone"}

print(f"Target Column: {target_column}")
print(f"Prediction Values: {target_values}")
```

## 3. Success metric

The success metric is Precision and Recall for the "Refresh" class, focusing specifically on outperforming the 24% accuracy of the legacy handwritten rule.

```python
# Defining our success metrics and baseline
success_metrics = ["Precision", "Recall"]
baseline_to_beat = 0.24  # 24% from the hand-written rule

print(f"Primary Metrics to Track: {success_metrics}")
print(f"Legacy Baseline to Beat: {baseline_to_beat:.0%}")
```

## 5. Why ML beats a fixed rule here

ML beats a fixed rule here because web performance decay is highly complex. A handwritten rule is too rigid to weigh the interactions between dozens of variables like age, current traffic, and click-through rates. An ML model can detect non-linear patterns across all these features, which allows it to reach a 74% success rate compared to the rule's 24%.

```python
# A simple programmatic check of expected improvement
rule_accuracy = 0.24
ml_accuracy = 0.74
improvement = ml_accuracy - rule_accuracy

print(f"Legacy Rule Accuracy: {rule_accuracy:.0%}")
print(f"Expected ML Accuracy: {ml_accuracy:.0%}")
print(f"Net Improvement with ML: +{improvement:.0%}")
```
